```python
# Title: An EDA of womens and mens college basketball from 2001 to 2024 (NCAA Season Averages) 
# Author: Yashvi Parmar
# Data: SportsReference 

import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'womensNCAA.txt'
delimiter = ','
womens_df = pd.read_csv(file_path, sep=delimiter)
```

```python
file_path = 'mensNCAA.txt'
delimiter = ','
mens_df = pd.read_csv(file_path, sep=delimiter)
```

```python
womens_df['year'] = womens_df['Season'].str.split('-').str[0]
womens_df['year'] = pd.to_numeric(womens_df['year'])
womens_df = womens_df[womens_df['year'] >= 2001]
```

```python
mens_df['year'] = mens_df['Season'].str.split('-').str[0]
mens_df['year'] = pd.to_numeric(mens_df['year'])
mens_df = mens_df[mens_df['year'] >= 2001]
```

```python
womens_df = womens_df[::-1]
mens_df = mens_df[::-1]
```

```python
plt.rcParams["font.family"] = "Courier New"
```

```python
stat = ['PTS', 'AST', 'FT', '3PA', '3P%', 'BLK', 'STL', 'TOV', 'TRB']
```

```python
def plot(stat):
    plt.figure(figsize=(20, 10))
    plt.title( stat + ' Over Seasons', fontweight="bold")
    plt.xticks(rotation=45)
    men_line, = plt.plot(mens_df['Season'], mens_df[stat])
    women_line, = plt.plot(womens_df['Season'], womens_df[stat])
    plt.xlabel('Season', fontweight="bold")
    plt.ylabel(stat, fontweight="bold")
    plt.legend([men_line, women_line], ['Men', 'Women']) 
    plt.show()
```

```python
#I created my own style sheet - sportsanalytics for the colors and design of the graphs
for i in range(len(stat)): 
    with plt.style.context('sportsanalytics'):
        plot(stat[i])
```
